```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('heart.csv')
print(df)
```

```python
df.info()
print('---------------------------------')
df.isna().sum()
print('---------------------------------')
df.head()
print('---------------------------------')
df.tail()
```

```python
# extracting x and y
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
print(x)
```

# Applying PCA 

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
x=pca.fit_transform(x)
print(pca.explained_variance_ratio_)
```

```python
print(x)
```

```python
# splitting x and y into training and test set
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr= LinearRegression()
lr.fit(x_tr,y_tr)
y_pr = lr.predict(x_te)
print(r2_score(y_te,y_pr))
```

```python
from sklearn.svm import SVC
cl=SVC(kernel='linear',random_state=0)
cl.fit(x_tr,y_tr)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
y_pr=cl.predict(x_te)
print(accuracy_score(y_pr,y_te))
```

```python
from mlxtend.plotting import plot_confusion_matrix,plot_decision_regions
cm=confusion_matrix(y_te,y_pr)
plot_confusion_matrix(cm)
plt.show()
```

```python
plot_decision_regions(X=x_te,y=y_te,clf=cl)
plt.show()
```
